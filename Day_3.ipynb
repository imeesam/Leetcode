```python
def big_nums(nums):
    temp = sorted(nums)
    d = {}
    for i,v in enumerate(temp):
        if v not in d:
            d[v]=i
    ret = []
    for i in nums:
        ret.append(d[i])
    return ret
    
```

```python
big_nums([8,1,2,2,3])
#  this is similar to two sum 
```

```python
def all_nums(nums):
    count = [0]*102
    for i in nums:
        count[i+1]+=1
    for i in range(1,102):
        count[i]+=count[i-1]
    return  [count[num] for num in nums]
```

# Problem 6

`the key to solve this problem is the minmum amount of moves we can is the max difference b/w x's and y's b/w two points`

```python

def min_time(points):
    res = 0
    x1,y1 = points.pop()
    while points:
        x2,y2 = points.pop()
        res += max(abs(y2-y1),abs(x2-x1))
        x1,y1 = x2,y2
    return res
```

```python
min_time([[1,1],[3,4],[-1,0]])
```

```python
def min_time(points):
    total = 0
    for i in range(len(points) - 1):
        total += max(
                abs(points[i][0] - points[i + 1][0]), 
                abs(points[i][1] - points[i + 1][1])
                )
    return total
```
