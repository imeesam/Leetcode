# Recursive method 

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
node15 = TreeNode(15)
node7 = TreeNode(7)
node20 = TreeNode(20, node15, node7)
node9 = TreeNode(9)
root = TreeNode(3, node9, node20)
```

```python
from collections import deque
def maxdepth(root):
    if not root:
        return 0

    return max(maxdepth(root.left),maxdepth(root.right))+1
```

```python
maxdepth(root)
```

```python
from collections import deque
def largestnode(root):
    queue = deque([root])
    max_node = 0
    while queue:
        curr_node =  queue.popleft()
        if curr_node.left:
            queue.append(curr_node.left)
        if curr_node.right:
            queue.append(curr_node.right)

        if curr_node.val > max_node:
            max_node = curr_node.val
    return max_node
```

```python
largestnode(root)
```

```python
def smallestnode(root):
    queue = deque([root])
    min_node = 1000
    while queue:
        curr_node =  queue.popleft()
        if curr_node.left:
            queue.append(curr_node.left)
        if curr_node.right:
            queue.append(curr_node.right)

        if curr_node.val < min_node:
            min_node = curr_node.val
    return min_node
```

```python
smallestnode(root)
```

```python
def levelnode(root):
    if root == None:
        return []
        
    queue =  deque([root])
    result = []

    while queue:
        level = []
        for i in range(len(queue)):
            node = queue.popleft()
            level.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        result.append(level)
    return result
```

```python
levelnode(root)
```

```python
nodel1 = TreeNode(1)
noder1 = TreeNode(2)
root1 = TreeNode(1,nodel1,noder1)

nodel2 = TreeNode(1)
noder2 = TreeNode(2)
root2 = TreeNode(1,nodel2,noder2)
```

```python
def sametree(p,q):
    stack =  [(p,q)]
    while stack:
        node1 , node2 = stack.pop()
        if not node1 and not node2:
            continue
        elif None in [node1 , node2] or node1.val != node2.val:
            return False

        stack.append((node1.right,node2.right))
        stack.append((node1.left,node2.left))

    return True
    
```

```python
sametree(root1,root2)
```
