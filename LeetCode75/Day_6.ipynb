```python
def stock_mark(array):
    l,r = 0,1
    maxprof = 0 
    while r != len(array):
        if array[l] < array[r]:
            prof =  array[r] - array[l]
            maxprof = max(maxprof,prof)
        else :
            l = r
        r+=1
    return maxprof
```

```python
stock_mark([7,1,5,3,6,4])
```

```python
# second solution 
def square_sortarray(array):
    if not array:
        return array

    if array[0]>0:
        return [i**2 for i in array]


    m = 0
    for i, n in enumerate(array):
        if n >= 0:
            m=i
            break
    A , B =  array[m:] , [-1*i for i in reversed(array[:m])]

    def merge(a,b):
        a = b = 0
        ret = []
        while a < len(A) and b < len(B):
            if A[a] < B[b]:
                ret.append(A[a])
                a+=1
            else:
                ret.append(B[b])
                b+=1

        if a<len(A):
            ret.extend(A[a:])
        else:
            ret.extend(B[b:])

        return [i**2 for i in ret]
    print(merge(A,B))
    
```

```python
square_sortarray([-2,-1,0,1,2,3,4])
```
