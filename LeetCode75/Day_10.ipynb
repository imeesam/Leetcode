```python
def climbStairs(n):
    if n<=2:
        return n
    else:
        dp = [0] * (n + 1)
        dp[0] = 0
        dp[1] = 1
        dp[2] = 2
        for i in range(3,n+1):
            dp[i] = dp[i-1] + dp[i-2]
        # print(dp)   
        return dp[n]
        
```

```python
climbStairs(6)
```

```python
def maxsubarray(nums):
    dp = [0]*len(nums)
    for i,n in enumerate(nums):
        dp[i] = max(dp[i], dp[i-1]+n)

    return max(dp)
```

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]
maxsubarray(nums)
```

```python
def maxsubarray(nums):
    maxsumm = nums[0]
    curr_sum = 0
    for i in nums:
        if curr_sum < 0:
            curr_sum = 0
        curr_sum +=i
        maxsumm = max(maxsumm,curr_sum)

    return maxsumm
```
