# Invert binary tree

doing this with depth first search and so here we have to use stack, if we go with the BFS then we have to use queue

```python
def invertbinarytree(root):
    stack =  [root]
    while stack:
        curr = stack.pop()
        if curr:
            curr.left , curr.right = curr.right , curr.left
            stack.extend([curr.right, curr.left])
    return root
```
