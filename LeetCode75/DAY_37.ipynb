is a data structure that is in a form binary tree and it's an ordered binary tree every node in left subtree is less than parent and every node in right subtree is greater than its parent
\
\
any node's left child value is less than it's parent value and node's right child value is greater than it's parent value

```python
def searchBST(root,val):
    while root:
        if root.value == val:
            return root
        elif root.val < val :
            root = root.right
        else:
            root = root.left
        
    return None
```
