```python
def time_k(tickets,k):
    result = 0
    for i in range(len(tickets)):
        if i <=k:
            result += min(tickets[i],tickets[k])
        else:
            result += min(tickets[i],tickets[k]-1)
    return result
```

```python
tickets = [5,5,2,2]
k = 2
```

```python
time_k(tickets,k)
```
