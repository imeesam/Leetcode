```python
nums = [2,3,1,2,4,3]
target = 7
```

```python
# use sliding array on it withy the help of two pointers 
```

```python
def miin(nums,target):
    l = 0
    min_diff =  float("inf")
    suum = 0
    for r in range(len(nums)):
        print('ith value',r)
        suum += nums[r]
        print('sum till ith value',suum)
        while suum>=target:
            min_diff = min(min_diff, r-l+1)
            print('minc diff at ith', r ,':', min_diff)
            suum -= nums[l]
            print('sum after update',suum)
            l+=1
            print('left_pointer' , l)
    
    if min_diff == float('inf'):
        return 0
    else:
        return min_diff
```

```python
miin(nums,target)
```

```python
n = 4
nums = [2,3,1,2]
for l in range(n) :
    for r in range(l, n):
        print('l =', l,'r =', r, 'r+1 = ', r+1, nums[l:r+1] , sum(nums[l:r+1]) )
```

# Bit Manupulation 

```python
a = [1,2,1,2,4]
xorr = 0
for i in a:
    xorr^=i
    print(xorr)
```

```python
num =  100
for bit in range(10):
    print(num>>bit)
```

```python
ones = 0 
twos = 0

for num in nums:
    twos |= ones & num
    ones ^= num
    mask = ~(ones&twos)
    ones &= mask
    twos &= mask
```
