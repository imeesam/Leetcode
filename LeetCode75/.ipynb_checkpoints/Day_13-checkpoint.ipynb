```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    def print_list(self):
        current = self
        while current:
            print(current.val, end=" -> ")
            current = current.next
        print("None")
```

```python
l5 = ListNode(5)
l4 = ListNode(4, l5)
l3 = ListNode(3, l4)
l2 = ListNode(2, l3)
l1 = ListNode(1, l2)
```

```python
l1.print_list()
```

```python
def reverse_ll(head):
    prev = None
    curr = head
    while(curr != None):
        next_pointer = curr.next
        curr.next = prev 
        prev = curr
        curr = next_pointer
    return prev
```

```python
new_head = reverse_ll(l1)
new_head.print_list()
```
