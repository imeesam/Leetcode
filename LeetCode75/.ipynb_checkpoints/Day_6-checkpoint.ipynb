# Problem 9

```python
def stock_mark(array):
    l,r = 0,1
    maxprof = 0 
    while r != len(array):
        if array[l] < array[r]:
            prof =  array[r] - array[l]
            maxprof = max(maxprof,prof)
        else :
            l = r
        r+=1
    return maxprof
```

```python
stock_mark([7,1,5,3,6,4])
```
