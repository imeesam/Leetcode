```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    def print_list(self):
        current = self
        while current:
            print(current.val, end=" -> ")
            current = current.next
        print("None")

```

```python
l5 = ListNode(5)
l4 = ListNode(4, l5)
l3 = ListNode(3, l4)
l2 = ListNode(2, l3)
l1 = ListNode(1, l2)
```

```python
l1.print_list()
```

```python
def middle_node(head):
    slow = fast = head
    while fast  and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow.print_list()
```

```python
middle_node(l1)
```
