```python
def removeDuplicateLetters(s):
    stack = []                         # This will store the result
    last = {c: i for i, c in enumerate(s)}  # Last occurrence of each char
    added = set()                      # Letters already in stack
    
    for i, c in enumerate(s):
        if c in added:                 # Skip if already in result
            continue
        
        # Remove letters that are bigger than current and appear later
        while stack and c < stack[-1] and i < last[stack[-1]]:
            removed = stack.pop()
            added.remove(removed)
        
        stack.append(c)
        added.add(c)
    print(last)
    return ''.join(stack)
```

```python
s = 'cbacdcbc'
removeDuplicateLetters(s)
```
