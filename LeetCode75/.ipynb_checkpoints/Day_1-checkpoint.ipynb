# Problem 1 

Can use 2 loops but the time complexity will be higher in it , 
\
\


A better solution  is to use set as it dosent allow to duplicates enter in it 
\
\


```python
def duplicate(nums):
    return len(set(nums)) != len(nums)
```

```python
duplicate([1,2,3,4])
```

# Problem 2

|

Initial soultion but this is very slow like this 
\
\


```python
def num(nums):
    nums.sort()
    for i,v in enumerate(nums):
        if i != v:
            return v-1
        if v == len(nums)-1:
            # if n = 3 then [range from 0-3], means 3 is missing from it 
            return v+1
```

```python
num([0,1,2]) 
```

Better Solution 


\
reason for this `+1` in `sum(nums)+1` is that range exclude's last value 
\
\


```python
def miss_number(nums):
    return sum(range(len(nums)+1)) - sum(nums)
```

```python
miss_number([9,6,4,2,3,7,0,1])
```
