```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
node15_left = TreeNode(15)
node7 = TreeNode(7)
node9 = TreeNode(9, node15_left, node7)
node20_right = TreeNode(20)
root = TreeNode(3, node7, node20_right)
```

```python
from collections import deque
```

```python
def avglevelofBT(root):
    queue = deque([root])
    result = []
    while queue:
        level = []
        for i in range(len(queue)):
            node = queue.popleft()
            level.append(node.val)
            if node.left : queue.append(node.left)
            if node.right: queue.append(node.right)
        result.append(sum(level)/len(level))
    return result
```

```python
avglevelofBT(root)
```
